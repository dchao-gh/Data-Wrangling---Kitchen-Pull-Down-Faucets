```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('ProductSummary.csv')
```

```python
kf = data[data[' Segment'] == 'Kitchen Faucets']
```

```python
pds = kf.loc[kf[' Description'].str.contains("Pull-Down", case=False)]
```

```python
name_change = pds.loc[:, ' Sales 2020W39' : ' Price 2021W38']
```

```python
new_column_names = []
new_names = ['SalesWeek', 'UnitsWeek', 'PriceWeek']
for i in new_names:
    for j in range(1,54,1):
        new_column_names.append(i+str(j))
```

```python
nn_df = name_change.set_axis(new_column_names, axis=1)
```

```python
units = nn_df.loc[:,'UnitsWeek1' : 'UnitsWeek53']
units = units.fillna(0)
```

```python
L = []
for (cN, cD) in units.iteritems():    
    cD_list = cD.to_list()
    new_List = []
    for value in cD_list:
        p = value/sum(cD_list)
        new_List.append(p)
    L.extend(new_List)

split = np.array_split(L, 53)
   
x_array = np.array(split)
x_trans = np.transpose(x_array)
```

```python
column_list = []
for i in range(1,54,1):
    column_list.append('PercentWeek' + str(i))
```

```python
import pandas as pd

percent_df = pd.DataFrame(data=x_trans, columns = column_list)
```

```python
new_col_names = nn_df.reset_index()
new_col_names = new_col_names.drop(columns = 'index')
spup = pd.concat([new_col_names, percent_df], axis=1)
```

```python
import pandas as pd
df1 = pds.loc[:, ' Site' : ' Sales Per Week Per Store']
df1 = df1.reset_index()
df1 = df1.drop(columns='index')
almost = pd.concat([df1,spup], axis=1)
```

```python
df2 = pds.loc[:, ' Certifications: ADA': ' Warranty']
df2 = df2.reset_index()
df2 = df2.drop(columns = 'index')
final = pd.concat([almost,df2], axis=1)
```

```python
pds.to_csv('pull_down_kitchen_faucets_only.csv')
final.to_csv('final_product_summary.csv')
```
